```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
```

```python
voc_size=10000
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
messege=X.copy()
```

```python
messege.reset_index(inplace=True)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
import re
```

```python
from nltk import corpus
```

```python
from nltk.corpus import stopwords
```

```python
messege.reset_index(inplace=True)
```

```python
ps=PorterStemmer()
corpes=[]
for i in range(0,len(messege)):
  review=re.sub('[^a-zA-Z]',' ', messege['title'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if word not in stopwords.words("english")]
  review=" ".join(review)
  corpes.append(review)


```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
one_words=[one_hot(words,voc_size) for words in corpes ]
```

```python
voc_size
```

```python
one_words
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
words_padd=pad_sequences(one_words,maxlen=20,padding="pre")
```

```python
print(words_padd)
```

```python
words_padd[0]
```

```python
len(words_padd)
```

```python
words_padd=np.array(words_padd)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(words_padd,y,train_size=0.75,random_state=100)
```

```python
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Embedding(voc_size,64,input_length=20))
model.add(Dense(1,activation='sigmoid'))
model.add(LSTM(100))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```
