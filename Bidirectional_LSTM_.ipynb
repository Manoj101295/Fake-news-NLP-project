```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Fake news/train.csv")
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
X=df.drop("label",axis=1)
```

```python
y=df.label
```

```python
import nltk
```

```python
import re
from nltk.stem import PorterStemmer
```

```python
le=PorterStemmer()
```

```python
text=X.copy()
```

```python
text.reset_index(inplace=True)
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
courpas=[]
for i in range(len(text)):
  review=re.sub('^[a-zA-Z]', ' ',text['title'][i])
  review=review.lower()
  review=review.split()
  review=[le.stem(word) for word in review if word not in stopwords.words('english')]
  review=' '.join(review)
  courpas.append(review)
```

```python
courpas
```

```python
import tensorflow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=5000
```

```python
onehot=[one_hot(word,voc_size) for word in courpas]
```

```python
onehot
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
length=20
onehot_pad=pad_sequences(onehot,maxlen=length)
```

```python
print(onehot_pad)
```

```python
len(onehot_pad)
```

```python
X=np.array(onehot_pad)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=.75,random_state=100)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Bidirectional
```

```python
model=Sequential()
model.add(Embedding(voc_size,64,input_length=length))
model.add(Bidirectional(LSTM(10)))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
pred=model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score


```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(accuracy_score(y_test,pred))
```
